```python
import pandas as pd
import numpy as np
low_memory=False
%matplotlib inline

# Dados dos candidatos a vereadores
# de São José dos Campos - SP 
# nas eleições de 2000

filename_prefeitos = 'eleicoes-brasil-2000_prefeitos.csv'
df_prefeitos = pd.read_csv(filename_prefeitos, encoding='utf-8', sep=',')
df_prefeitos.sample(10)
```

```python
data = {
    'Sigla do Partido': df_prefeitos['sigla_partido'],
    'Nome do Candidato': df_prefeitos['nome_candidato'],
    'Todal de Votos': df_prefeitos['total_votos'],
    'Resultado': df_prefeitos['desc_sit_cand_tot']
}

new_df_prefeitos = pd.DataFrame(data=data)
new_df_prefeitos
```

```python
new_df_prefeitos.groupby('Sigla do Partido').sum('Total de Votos').plot(
    kind='barh',
    title="Número de votos por Partido"
)
```
