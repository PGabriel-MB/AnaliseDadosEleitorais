```python
import pandas as pd
import numpy as np
low_memory=False
%matplotlib inline

# Dados dos candidatos a VEREADORES
# de São José dos Campos - SP 
# nas eleições de 2000

filename_vereadores = 'eleicoes-brasil-2000_vereadores.csv'
df_vereadores = pd.read_csv(filename_vereadores, encoding='utf-8', sep=',')
df_vereadores.sample(20)
```

```python
data = {
    'Sigla do Partido': df_vereadores['sigla_partido'],
    'Nome do Candidato': df_vereadores['nome_candidato'],
    'Todal de Votos': df_vereadores['total_votos'],
    'Resultado': df_vereadores['desc_sit_cand_tot']
}

new_df_vereadores = pd.DataFrame(data=data)
new_df_vereadores.groupby('Sigla do Partido').sum('Total de Votos')
```

```python
new_df_vereadores.groupby('Sigla do Partido').sum('Total de Votos').plot(
    kind='barh',
    title="Número de votos por Partido - 2000"
)
```

```python
data = {
    'Nome do Candidato': df_vereadores['nome_candidato'],
    'Todal de Votos': df_vereadores['total_votos'],
    'Resultado': df_vereadores['desc_sit_cand_tot']
}

new_df_vereadores = pd.DataFrame(data=data)
new_df_vereadores.groupby('Resultado').sum('Total de Votos').plot(
    kind='bar',
    title="Número de votos por Candidato a VERADORES - 2000"
)
```

```python
data = {
    'Nome do Candidato': df_vereadores['nome_candidato'],
    'Total de Votos': df_vereadores['total_votos'],
    'Resultado': df_vereadores['desc_sit_cand_tot']
}

new_df_vereadores = pd.DataFrame(data=data)
new_df_vereadores = new_df_vereadores[new_df_vereadores['Resultado'] == 'ELEITO']
new_df_vereadores.groupby('Nome do Candidato').sum('Total de Votos').plot(
    kind='barh',
    title='Total de votos dos VEREADORES ELEITOS - 2000'
)
```
